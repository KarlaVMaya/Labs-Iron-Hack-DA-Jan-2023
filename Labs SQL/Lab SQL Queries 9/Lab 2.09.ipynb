```python
def set_month_return_df(month,engine):
    query="SELECT customer.first_name, customer.last_name, a.total_rent_{month} FROM customer \
LEFT JOIN( SELECT rentals_{month}.customer_id, COUNT(rentals_{month}.customer_id) AS total_rent_{month} FROM rentals_{month} \
LEFT JOIN customer ON (rentals_{month}.customer_id=customer.customer_id) GROUP BY 1) AS a ON(customer.customer_id=a.customer_id) ORDER BY a.total_rent_{month} DESC".format(month=month)
    data_rentals= pd.read_sql_query(query, engine)
    return data_rentals
```

```python
df_june=set_month_return_df("june",engine)
df_june
```

```python
df_may=set_month_return_df("may",engine)
df_may
```

```python
def concat_df_from_months(month1,month2,engine):
    df_m1=set_month_return_df(month1,engine)
    df_m2=set_month_return_df(month2,engine)
    concat_df= pd.concat([df_m1, df_m2], axis=1)
    return concat_df
```
