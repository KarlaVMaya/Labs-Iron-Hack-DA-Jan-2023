```python
#Load data from the merged_clean_ver2.csv file
data1= pd.read_csv("merged_clean_ver2.csv")
data1
```

```python
data1=data1.drop(columns=['Unnamed: 0'])
data1=data1
data1
```

```python
import matplotlib.pyplot as plt
plt.plot(data1['ic2'],data1['ic3'],'o')
```

```python
#Plot Pearson Correlation Heatmap.
import scipy.stats as stats
data_num=data1.drop(['state', 'gender','ic2_'],axis=1)
data_num
```

```python
type(data_num)
```

```python
corr_matrix=data_num.corr()
corr_matrix
```

```python
import seaborn as sns 
mascara=np.zeros_like(corr_matrix)
mascara[np.triu_indices_from(mascara) ]=True 

fig , ax=plt.subplots(figsize=(10,8))
ax= sns.heatmap(corr_matrix, mask=mascara, annot=True, cmap='RdYlGn', xticklabels=['id'])
```

```python
#Plot Spearman Correlation Heatmap.
corr_matrix=data_num.corr(method='spearman')
corr_matrix
```

```python
mascara=np.zeros_like(corr_matrix)
mascara[np.triu_indices_from(mascara) ]=True 

fig , ax=plt.subplots(figsize=(10,8))
ax= sns.heatmap(corr_matrix, mask=mascara, annot=True, cmap='coolwarm', xticklabels=['id'])
```

```python
corr_matrix=data_num.corr(method='kendall')
corr_matrix
```

```python
mascara=np.zeros_like(corr_matrix)
mascara[np.triu_indices_from(mascara) ]=True 

fig , ax=plt.subplots(figsize=(10,8))
ax= sns.heatmap(corr_matrix, mask=mascara, annot=True, cmap='BuPu', xticklabels=['id'])
```
