```python
#Import the necessary libraries if you are starting a new notebook.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the csv. Use the variable customer_df as customer_df = pd.read_csv()
customer_df=pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
customer_df.columns=[c.lower().replace(' ', '_') for c in customer_df.columns]
customer_df.head()
```

```python
#Get the numerical variables from our dataset.
data_num= customer_df.select_dtypes(include=["float64","int64"])
data_num
```

```python
data_num.isnull().sum()
```

```python
#Check using a distribution plot if the variables fit the theoretical 
#normal or exponential distribution.

for c in data_num:
    sns.distplot(data_num[c])
    plt.show()
```
